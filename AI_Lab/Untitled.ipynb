```python
def dfs(graph,start,end, visited = None, path = None, path_cost=0):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path.append(start)

    if start == end:
        print(path, path_cost)
        return True

    for neighbour, weight in graph[start]:
        if neighbour not in visited:
            if dfs(graph, neighbour, end, visited, path, path_cost+weight):
                return True
    path.pop()
    visited.remove(start)
    return False
```

```python
from collections import deque
def bfs(graph, start, end):
    visited = set()
    queue = deque([(start,0)])
    parent = {start:None}

    while queue:
        node, cum_weight = queue.popleft()

        if node == end:
            path = []
            while node is not None:
                path.append(node)
                node = parent[node]
            return path[::-1], cum_weight

        if node not in visited:
            visited.add(node)
            for neighbour, weight in graph[node]:
                if neighbour not in visited:
                    queue.append((neighbour, cum_weight+weight))
                    parent[neighbour] = node
    return None
        
```

```python
import heapq

def ucs(graph, start, goal):
    priority_queue = [(0,start,[start])]
    visited = set()

    while priority_queue:
        cost, node, path = heapq.heappop(priority_queue)

        if node not in visited:
            continue

        visited.add(node)

        if node == goal:
            return cost, path

        for neighbour, weight in graph[node]:
            if neighbour not in visited:
                heapq.heappush(priority_queue, (cost+weight, neighbour, path + [neighbour]))
    return float('inf'), []
```
